## Data Manipulation

```python
countries_of_interest = ['Canada', 'Finland', 'Italy', 'Kenya', 'Norway', 'Singapore']
filtered_data = df[df['Country Name'].isin(countries_of_interest)]

# Select data for the years 2010 to 2019
years_of_interest = [str(year) for year in range(2010, 2020)]
filtered_data = filtered_data[['Country Name'] + years_of_interest]

# Reset index for clarity
filtered_data.reset_index(drop=True, inplace=True)
filtered_data.set_index('Country Name', inplace=True)

transposed_data = filtered_data.T
```

```python
display(filtered_data)
display(transposed_data)
```

```python
#Rename the columns
transposed_data.columns = ['GDP_' + col for col in transposed_data.columns]

# Reset index so that 'Year' becomes a column
transposed_data.reset_index(inplace=True)
transposed_data.rename(columns={'index': 'year'}, inplace=True)

transposed_data
```

```python
transposed_data.to_csv('filtered_gdp_per_capita.csv', index=False)
```
